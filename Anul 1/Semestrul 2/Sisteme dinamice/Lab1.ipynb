```python
#8a
t=var('t')
f(t,s)=1-(s*cos(4*t)*cos(t))/sqrt(1-s^2*(cos(4*t))^2*(sin(t))^2)
x(t)=f(t-(pi/2),0.5)
y(t)=f(t,0.5)
parametric_plot((x(t),y(t)),(t,0,2*pi))
```

```python
#8b
t,s=var('t,s')
f(t,s)=1-(s*cos(4*t)*cos(t))/sqrt(1-s^2*(cos(4*t))^2*(sin(t))^2)
x(t)=f(t-(pi/2),s)
y(t)=f(t,s)
g=parametric_plot((x(t).subs(s=0.1), y(t).subs(s=0.1)),(t,0,2*pi))
for k in[2..10]:
    g1=parametric_plot((x(t).subs(s=k/10),y(t).subs(s=k/10)),(t,0,2*pi))
    g=g+g1
show(g)
```
