```python
#ex1a
t,k,w0,v0,x0=var('t,k,w0,v0,x0')
x=function('x')(t)
deq = diff(x,t,2) + w0^2*x == 0
assume(w0>0)
sol = desolve(deq,[x,t])
show(sol)
```

```python
#1b
soll = desolve(deq,[x,t], ics=[0, x0, v0])
show(soll)
```

```python
#1c
L=0.3924
g=9.81
w0=sqrt(g/L) #frecventa
x0=0.15
v0=0
R=sqrt(x0^2+(v0/w0)^2)  #R-amplitudinea  
T=(2*pi/w0) #T- perioada
d=0 #d-faza miscarii

z=function('x')(t)
z=R*cos(w0*t-d)
show(z(t).full_simplify())
plot(z(t),t,-T,T)
```

```python
#2a
reset()
t,w0,l=var('t,w0,l')
x=function('x')(t)
deq = diff(diff(x,t),t) + l*diff(x,t) + w0^2*x == 0
assume(w0,'integer')
assume(l^2>4*w0^2)
sol = desolve(deq,[x,t])
show(sol)
```

```python
#2b
sol1=desolve(deq,[x,t],ics=[0,1,5])
sol2=sol1.substitute(l==25,w0==10)
show(sol2)
plot(sol2,0,2)
```

```python
#2c
reset()
t,w0,l=var('t,w0,l')
x=function('x')(t)
assume(l,'integer')
assume(l>0)
deq = diff(diff(x,t),t) + l*diff(x,t) + (l^2/4)*x == 0
sol = desolve(deq,[x,t])
show(sol)
```

```python
#2d
sol1=desolve(deq,[x,t],ics=[0,1,5])
sol2=sol1.substitute(l==20)
show(sol2)
plot(sol2,0,2)
```

```python
#3a
reset()
w,w0,F0,t=var('w,w0,F0,t')
x=function('x')(t)
deq = diff(x,t,2) + w0^2*x == F0*cos(w*t)
assume(w0>0)
assume(w0!=w)
sol=desolve(deq,[x,t])
show(sol)
```

```python
#3b
sol1=desolve(deq,[x,t],ics=[0,0,0])
sol2=sol1.substitute(w0==5, w==5.5, F0==2)
show(sol2)
plot(sol2,t,0,4,ymin=-2,ymax=2)
```

```python
#3c
reset()
w,w0,F0,t=var('w,w0,F0,t')
x=function('x')(t)
deq = diff(x,t,2) + w0^2*x == F0*cos(w0*t)
assume(w0>0)
sol=desolve(deq,[x,t])
show(sol)
```

```python
#3d
sol1=desolve(deq,[x,t],ics=[0,0,0])
sol2=sol1.substitute(w0==5, w==5, F0==2)
show(sol2)
plot(sol2,t,0,4,ymin=-2,ymax=2)
```

```python
#3e
reset()
w,w0,F0,t=var('w,w0,F0,t')
x=function('x')(t)
deq = diff(x,t,2) + w0^2*x == F0*cos(w0*t)
assume(w0>0)
sol=desolve(deq,[x,t],ics=[0,0,0])
show(sol)
show(limit(sol(t,w),w=w0))
show(sol(t,w0).full_simplify())
```
