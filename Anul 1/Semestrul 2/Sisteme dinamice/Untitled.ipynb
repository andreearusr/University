```python
#1a
reset()
x=var('x')
y=function('y')(x)
eqd=diff(y,x)==2*x*(1+y^2)
sol = desolve(eqd,y)
show(sol)
solve(sol,y)
```

```python
reset()
x=var('x')
y=function('y')(x)
eqd=diff(y,x)==2*x*(1 + y^2)
show(desolve(eqd,y))
sol = desolve(eqd,y)
_C=var('_C')
[sol1]=solve(sol,y(x))
show(sol1)
f(x,_C) = sol1.rhs()
plot(f(x,0),x,-10,10,ymin=-10,ymax=10)

```

```python
#2a
reset()
x=var('x')
y=function('y')(x)
eqd=diff(y,x)==1+y^2
sol = desolve(eqd,y,ics=[0,1])
show(sol)
[sol1]=solve(sol,y(x))
show(sol1)
f(x) = sol1.rhs()
plot(f(x),x,-10,10,ymin=-5,ymax=5)
```

```python
#2d
reset()
x=var('x')
y=function('y')(x)
eqd=diff(diff(y,x))-5*diff(y,x)+4*y==0
sol = desolve(eqd,y,ics=[0,5,8])
show(sol)
plot(sol,x,-2,2)
```

```python
#3a
reset()
x,k=var('x,k')
y=function('y')(x)
eqd=diff(y,x)+k*y/x==x^3
sol=desolve(eqd,y,ivar=x)
show(sol)
```

```python
#3b
reset()
x=var('x')
y=function('y')(x)
eqd=diff(y,x)+y/x==x^3
sol=desolve(eqd,y)
show(sol)
_C=var('_C')
plot(sol.subs(_C==2),-2,2,ymin=-5,ymax=5)
```

```python
#3c
reset()
x=var('x')
y=function('y')(x)
eqd=diff(y,x)+y/x==x^3
sol = desolve(eqd,y,ics=[1,0])
show(sol)
plot(sol,x,-2,2,ymin=-5,ymax=5)
```

```python
#3d
reset()
x,k=var('x,k')
y=function('y')(x)
eqd=diff(y,x)+k*y/x==x^3
sol(x,k)=desolve(eqd,y,ivar=x,ics=[1,0])
show(sol(x,k))
p=[plot(sol(x,k), x, -2, 2, ymin=-10, ymax=10) for k in [5..10]]
animate(p)
```
