```python
#3b
reset()
x=var('x')
k=1
y=function('y')(x)
eqd=diff(y,x)+(k*y)/x==x^3
sol=desolve(eqd,[y,x])
show(sol)
_C=var('_C')
sol1=sol.substitute(_C==2)
sol2=sol.substitute(_C==-3)
sol3=sol.substitute(_C==5)
plot([sol1,sol2,sol3],-2,2,ymin=-5,ymax=5)
```
