```python
#1
reset()
t,r=var('t,r')

x=function('x')(t)
eq=diff(x,t)==r*x*(1-x)
sol=desolve(eq,[x,t])
show(sol)
show(solve(sol,x))
```

```python
#2
reset()
t=var('t')
x=function('x')(t)
deq=diff(x,t)==x*(1-x)

g1=desolve_rk4(deq,x,[0,0.1], step=0.5, end_points=[0,5], output='plot')
g2=desolve_rk4(deq,x,[0,2], step=0.5, end_points=[0,5], output='plot', color='red')

show(g1+g2)
```

```python
#3
s=var('s')
f(s)=s*(1-s)
g3=plot_slope_field(f(s), (t,0,-5), (s,-1,2), headaxislength=3, headlength=4, color='red')
show(g1+g2+g3)
```

```python
#4
reset()
s=var('s')
f(s)=s*(1-s)*(2-s)
eq=solve(f(s)==0,s)
show(eq)

s1=eq[0].rhs()
s2=eq[1].rhs()
s3=eq[2].rhs()

show(diff(f,s)(s1))  #s1=0  local as stabil
show(diff(f,s)(s2))  #s2=1  instabil
show(diff(f,s)(s3))  #s3=2  local as stabil
```
