```python
#problema 2a
reset()
t=var('t')
x=function('x')(t)
f=function('s')(t)
f(s)=s*(3-s^2)
eqp=solve(f(s)==0,s)
eqp
```

```python
s1=eqp[0].rhs()
diff(f,s)(s1)  #s1 punct de echilibru local asimptotic stabil
```

```python
s2=eqp[1].rhs()
diff(f,s)(s2)  #s1 punct de echilibru local asimptotic stabil
```

```python
s3=eqp[2].rhs()
diff(f,s)(s3)  #s1 punct de echilibru instabil
```
